#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def preproc_data(df_input):
    df_output = df_input.copy()
    
    ## Your Code Here
    # Удаляло идентификатор `AGREEMENT_RK`
    df_output = df_output.drop(['AGREEMENT_RK'], axis=1)

    # Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
    for i in ['PERSONAL_INCOME', 'CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT']:
        df_output[i] = df_output[i].map(lambda x: x.replace(',', '.')).astype('float')

    # Что-то делала с пропусками
    df_output = df_output.fillna(value={'PREVIOUS_CARD_NUM_UTILIZED': 0, 'WORK_TIME': 0})
    
    # Кодировала категориальные признаки
    cat_cols = ['EDUCATION', 'MARITAL_STATUS', 'GEN_INDUSTRY', 'GEN_TITLE', 'ORG_TP_STATE', 'ORG_TP_FCAPITAL', 'JOB_DIR', 'FAMILY_INCOME', 'REG_ADDRESS_PROVINCE', 'FACT_ADDRESS_PROVINCE','POSTAL_ADDRESS_PROVINCE', 'TP_PROVINCE','REGION_NM']
    df_output = pd.get_dummies(df_output, columns=cat_cols, prefix=cat_cols, dummy_na=True)
    #for i in cat_cols:
        #df_output[i] = df_output[i].astype('category').cat.codes
        
        
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)

df_train_preproc = df_preproc.query('sample == "train"').drop(['sample'], axis=1)
df_test_preproc = df_preproc.query('sample == "test"').drop(['sample'], axis=1)
```

#### Задание 4. Отделите целевую переменную и остальные признаки

Должно получится:
* 2 матрицы: X и X_test
* 2 вектора: y и y_test

```python
y = df_train_preproc['TARGET']
X = df_train_preproc.drop(['TARGET'], axis=1)
y_test = df_test_preproc['TARGET']
X_test = df_test_preproc.drop(['TARGET'], axis=1)
```

#### Задание 5. Обучение и оценка качества разных моделей

```python
from sklearn.model_selection import train_test_split
# test_size=0.3, random_state=42

## Your Code Here
#X, X_test = train_test_split(df_test_preproc, test_size=0.33, random_state=42)


# ------> вот тут я не совсем понял надо ли мне Х и y разделить на обучение и тест или для теста сразу можно использовать X_test и y_test, коль уж мы их знаем... будем считать, что можно 
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
# Постепенно мы узнаем, что они делают а сейчас учимся понимать какой работает качественнее
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

## Your Code Here
models = []
models.append({'model': DecisionTreeClassifier(), 'predict': 0, 'name': 'DTC'})
models.append({'model': RandomForestClassifier(), 'predict': 0, 'name': 'RF'})
models.append({'model': LogisticRegression(penalty='l1', solver='liblinear'), 'predict': 0, 'name': 'LR'})
for m in models:
    m['model'].fit(X, y)
    m['predict'] = m['model'].predict(X_test)
    m['predict_probe'] = m['model'].predict_proba(X_test)
 
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall

from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve

## Your Code Here
data = []
for m in models:
    data.append({
        'name': m['name'], 
        'precision': precision_score(y_test, m['predict']), 
        'recall': recall_score(y_test, m['predict']), 
        'accuracy': accuracy_score(y_test, m['predict']), 
        'roc_auc': roc_auc_score(y_test, m['predict_probe'][:,1]),
    })
    
data_df = pd.DataFrame(data)

data_df
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально сравнить)
# Возможно вас удивит качество! Но задача подобрана специально ;) Такое качество тоже бывает

## Your Code Here
%matplotlib inline
from matplotlib import pyplot as plt


_, ax_here = plt.subplots(figsize=(20,10))
data_df.plot(logy=False, x='name', kind='bar', stacked=False, ax=ax_here)
plt.title("DTC vs RF vs LR")
plt.show()
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе
# Подпишите оси и линии

## Your Code Here
from sklearn.metrics import roc_auc_score, roc_curve

plt.figure(figsize=(5, 5))
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
plt.plot([0, 1], [0, 1])

for m in models:
    fpr, tpr, thresholds = roc_curve(y_test, m['predict_probe'][:,1])
    plt.plot(fpr, tpr)
    
```

```python
from sklearn.cross_validation import cross_val_score
# Сделать k-fold (10 фолдов) кросс-валидацию каждой модели
# И посчитать средний roc_auc

## Your Code Here
cross_val_result = {}

for m in models:
    cross_val_result[m['name']] = cross_val_score(
    m['model'],
    X,
    y,
    scoring='roc_auc',
    cv=10)
```

```python
cvr_df = pd.DataFrame(cross_val_result)

_, ax_here = plt.subplots(figsize=(20,10))
cvr_df.plot(kind='line', ax=ax_here)
plt.title("DTC vs RF vs LR - ROC AUC")
plt.show()
```

```python
# Взять лучшую модель и сделать предсказания (с вероятностями (!!!)) для test выборки


## Your Code Here
best_model = 2 #LR
best_model_info = models[best_model]
predict = best_model_info['predict_probe']
predict_df = pd.DataFrame(y_test)
predict_df['PREDICT'] = predict[:,1]
predict_df['PREDICT_VAL'] = best_model_info['predict']
#predict_df[predict_df['PREDICT'] > 0.5]
#predict_df[(predict_df['PREDICT_VAL'] == predict_df['TARGET']) & ( predict_df['TARGET'] == 1)]
```

```python
# Померить roc_auc на тесте
# Вывести текстом и на графике =)

## Your Code Here
print('ROC AUC for ' + best_model_info['name'] + ': ', roc_auc_score(y_test, best_model_info['predict_probe'][:,1]))

plt.figure(figsize=(5, 5))
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve ' + best_model_info['name'])
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
plt.plot([0, 1], [0, 1])

fpr, tpr, thresholds = roc_curve(y_test, best_model_info['predict_probe'][:,1])
plt.plot(fpr, tpr)
```
